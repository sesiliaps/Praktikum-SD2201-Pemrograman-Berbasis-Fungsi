```python
import pandas as pd
from functools import reduce

# Baca file XLSX
data = pd.read_excel("/content/datasetpbf.xlsx")

# Misalnya kita memiliki kolom 'penjualan' di DataFrame
penjualan = data['Penjualan']

# Menggunakan reduce dan lambda untuk menghitung total penjualan
total_penjualan = reduce(lambda x, y: x + y, penjualan)

print("Total Penjualan:", total_penjualan)
```

```python
from functools import reduce

# Fungsi untuk menjumlahkan total penjualan
total_harga = lambda x, y : x + y

# Contoh data penjualan toko
toko_alfamart = {
    'barang': ["shampoo", "sabun", "mie", "pasta gigi"],
    'harga': [23000, 3500, 3100, 4000],
    'beratbersih': ["170 ml", "65 gr", "85 gr", "45 gr"]
}

# Menggabungkan semua list menjadi satu
hasil_gabungan = reduce(lambda x, y: x + y, [toko_alfamart['barang'], toko_alfamart['harga'], toko_alfamart['beratbersih']])
print("Gabungan list :", hasil_gabungan)

# Menggunakan reduce untuk menghitung total penjualan
total_semua_harga = reduce(total_harga, toko_alfamart['harga'])
print("Total penjualan dari barang toko:", total_semua_harga)
```
